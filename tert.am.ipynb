```python
url = "https://staff.am/en/jobs"
page = requests.get("https://staff.am/en/jobs")
response = TextResponse(url = page.url, body = page.text, encoding = "utf-8")
```

```python
def job_names(url):
    page = requests.get(url) #get the page
    response = TextResponse(url = page.url, body = page.text, encoding = "utf-8") #diferenciate html and text
    jobs = response.css('p[class="font_bold"]::text').extract()
    return jobs
```

```python
all_jobs = []
i=1
while True:
    url="https://staff.am/en/jobs?page={}&per-page=50".format(i)
    current_page_jobs = job_names(url)
    if len(current_page_jobs)==0:
        break
    else:
        all_jobs.extend(current_page_jobs)
        i=i+1
        time.sleep(2)
       
```

```python
response.css('p[class="job_list_company_title"]::text').extract()
```

```python
def company_names(url):
    page = requests.get(url) #get the page
    response = TextResponse(url = page.url, body = page.text, encoding = "utf-8") #diferenciate html and text
    company = response.css('p[class="job_list_company_title"]::text').extract()
    return company
```

```python
all_company = []
i=1
while True:
    url="https://staff.am/en/jobs?page={}&per-page=50".format(i)
    current_page_company = company_names(url)
    if len(current_page_company)==0:
        break
    else:
        all_company.extend(current_page_company)
        i=i+1
        time.sleep(2)
       
```

```python
def most_frequent(all_company):       #a) Which company has the highest number of job postings?
    counter = 0                    
    num = all_company[0] 
      
    for i in all_company: 
        curr_frequency = all_company.count(i) 
        if(curr_frequency> counter): 
            counter = curr_frequency 
            num = i 
  
    return num 

print(most_frequent(all_company))
```
